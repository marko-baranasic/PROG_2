```python
grades = [("Alice", 85), ("Bob", 92), ("Charlie", 78), ("David", 90), ("Eve", 50), ("Grace", 34)]
print("Original grades: ", grades)
```

```python
sorted_grades = sorted(grades)
print("Grades sorted by name: ", sorted_grades)
```

```python
sorted_grades_value = sorted(grades, key=lambda x: x[1])
print("Grades sorted by value: ", sorted_grades_value)
```

```python
names = ["Alice", "Bob", "Charlie", "David", "Eve", "Gracce", "Bruno", "AAA"]
upper_all = map(str.upper, names)
print(upper_all)
```

```python
upper_all = list(upper_all)
print(upper_all)
```

```python
ascii_val = [[ord(char) for char in s] for s in names]
print(ascii_val)

ascii_val_sum = [sum(ord(char) for char in s) for s in names]
print(ascii_val_sum)


####################
ascii_names = [list(map(ord, name)) for name in names]
sum_names = list(map(sum, ascii_names))
print(ascii_names)
```

```python
sums_names = list(map(sum, map(lambda name: list(map(ord, name)), names)))
print(sum_names)
```

```python
ascii_names2 = list(map(lambda x:sum(ord(c) for c inx), names))
print(ascii_names2)
```
