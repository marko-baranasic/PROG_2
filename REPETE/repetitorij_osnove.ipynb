```python
ukupno_sekundi = int(input("Unesi broj sekundi: "))


sati = ukupno_sekundi // 3600
ostatak_nakon_sati = ukupno_sekundi % 3600

minute = ostatak_nakon_sati // 60
sekunde = ostatak_nakon_sati % 60

print(f"Rezultat: {sati} h, {minute} min, {sekunde} s")
```
