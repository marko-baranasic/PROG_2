```python
def num_between(donja, gornja, broj):
    if broj > donja and broj < gornja:
        print("broj je unutar raspona")    
    elif broj < donja:
        print("broj je manji od raspona")
    elif broj > gornja:
        print("broj je veći od raspona")
```

```python
user_input = input("Unesi string: ")

for char in user_input:
    if char != " " and not char.isalpha():
        break
    elif char == " ":
        continue
    elif char.isalpha():
        print(char)
    
```
